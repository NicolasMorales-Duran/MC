```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
arc = np.genfromtxt("running_dirty_nov.gpx",delimiter=',') #Importamos el archivo ya procesado 
lt=arc[:,0]
lg=arc[:,1]
alt=arc[:,2]
hour=arc[:,3]
mins=arc[:,4]
sec=arc[:,5]
```

```python
R =  6378.1 #El radio de la tierra
x = R*np.cos(lt*np.pi/180.0)*np.cos(lg*np.pi/180.0) #Hacemos el cambio a coordenadas cartesianas
y = R*np.cos(lt*np.pi/180.0)*np.sin(lg*np.pi/180.0)
z = R*np.sin(lt*np.pi/180.0)
t= hour+mins/60+sec/3600 #Ponemos el tiempo en horas
```

```python
xin = x - x[0] #Definimos las posiciones y tiempo iniciales
yin = y - y[0]
zin = z - z[0]
tin = t-t[0]

latbog = 4.5981 #Latitud de Bogotá
longbog = -74.0758 #Longitud de Bogotá

#Definimos los dos vectores como se muestra en el link del sistema de coordenadas esférico

rxthe = -np.sin(longbog*np.pi/180.0) 
rythe = np.cos(longbog*np.pi/180.0)
rzthe = 0.0

rxphi = -np.cos(longbog*np.pi/180.0)*np.sin(latbog*np.pi/180.0)
ryphi = -np.sin(longbog*np.pi/180.0)*np.sin(latbog*np.pi/180.0)
rzphi = np.cos(latbog*np.pi/180.0)

#Definimos los vectores de desplazamiento en el plano tangente a Bogota

unit1 = [xin[i]*rxthe + yin[i]*rythe + zin[i]*rxthe for i in range(len(xin))]

unit2 = [xin[i]*rxphi + yin[i]*ryphi + zin[i]*rxphi for i in range(len(xin))]
```

```python
#Hacemos la gráfica de la trayectoria
plt.figure(figsize=(20,20))
plt.plot(unit1,unit2)
plt.show()
```

```python
#Integracion numérica de la distancia recorrida
ini = [(np.sqrt((unit1[i+1] - unit1[i])**2) + ((unit2[i+1]-unit2[i])**2)) for i in range(0,len(unit1) -1 ,1)] 
ini.insert(0,0)
ini1 = np.cumsum(ini)
```

```python
plt.figure(figsize=(20,20)) #Hacemos la gráfica de la distancia recorrida
plt.plot(tin,ini1)
plt.show()
```
