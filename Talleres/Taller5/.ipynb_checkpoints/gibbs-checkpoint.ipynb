```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
import itertools
from scipy.optimize import basinhopping
```

```python
fig=plt.figure(figsize=(15,10)) #Graficamos los primeros trminos de la serie de Fourier de la funcin ventana de altura \frac{\pi}{4}, se puede apreciar cmo se va aproximado cada vez ms la serie.
x = np.linspace(0.0,2*np.pi,128)
y=0.*x
plt.plot([0.,0.,np.pi,np.pi,2*np.pi,2*np.pi],[0.,np.pi/4,np.pi/4,-np.pi/4,-np.pi/4,0.],"k",lw=3)
for i in range(5):
    y+=np.pi/np.pi*1./(2*i+1)*np.sin((2*i+1)*x) 
    plt.plot(x,y) 
plt.show()
```

```python
def f(x,it):#Definimos la funcin que devuelve el mnimo de la serie de Fourier para un orden determinado
    y=0.*x
    for i in range(it):
        y+=np.pi/np.pi*1./(2*i+1)*np.sin((2*i+1)*x)
    return y
minimizer_kwargs = {"method": "BFGS"}
y = []
n=15

for i in range(n):
    def g(x):
        return f(x,i)
    ret = basinhopping(g, np.pi, minimizer_kwargs=minimizer_kwargs,niter=128)
    y.append(abs(ret.fun))
    
T = []
for i in range(1,len(y[1:])): #Hacemos la grfica de la constante de gibbs para diversos rdenes de la aproximacin de Fourier, vemos que el resultado tiende al valor terico que se encuentra en la Wikipedia.
    T.append((2*y[1:][i]-np.pi/2)/np.pi) 
        
plt.figure(figsize=(15,10))
plt.plot(range(1,len(y[1:])),T)
plt.plot(range(1,len(y[1:])),T,color='red')
plt.xlabel('order')
plt.ylabel(r'$(G - \pi/2)/\pi$')
plt.show()
```
