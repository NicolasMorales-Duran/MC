```python
%pylab
from matplotlib import animation
```

```python
choreography=genfromtxt("Rayleigh.csv",delimiter=",")
x=choreography[:,0]
y=choreography[:,1]
xdot=choreography[:,2]
ydot=choreography[:,3]
```

```python
len(x)/1600
```

```python
%matplotlib osx
fig=figure()
ax = fig.add_subplot(111, autoscale_on=False, xlim=(-36, 36), ylim=(-36, 36))
ax.grid()
line, = ax.plot([], [], 'o', lw=2)

def init():
    line.set_data([], [])
    return line

def animate(i):
    thisx = [x[i]]
    thisy = [y[i]]
    line.set_data(thisx, thisy)
    return line

ani = animation.FuncAnimation(fig, animate, range(250),
    interval=25, blit=False, init_func=init)

show()
```
